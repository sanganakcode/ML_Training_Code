```python
from google.cloud import vision

def detect_labels_uri(uri):
    """Detects labels in the file located in Google Cloud Storage or on the Web."""
    client = vision.ImageAnnotatorClient()
    image = vision.Image()
    image.source.image_uri = uri

    response = client.label_detection(image=image)
    labels = response.label_annotations
    return [label.description for label in labels]

# Test with a sample image
sample_image_uri = "https://wallpapercave.com/wp/wp3608722.jpg"  # Replace with any public image URL
labels = detect_labels_uri(sample_image_uri)
print(labels)
```

```python
from google.cloud import vision

def analyze_image(uri):
    client = vision.ImageAnnotatorClient()
    image = vision.Image()
    image.source.image_uri = uri

    # Perform label detection
    label_response = client.label_detection(image=image)
    labels = [label.description for label in label_response.label_annotations]

    # Perform text detection (OCR)
    text_response = client.text_detection(image=image)
    texts = [text.description for text in text_response.text_annotations]

    # Perform face detection
    face_response = client.face_detection(image=image)
    faces = face_response.face_annotations

    # Perform landmark detection
    landmark_response = client.landmark_detection(image=image)
    landmarks = [landmark.description for landmark in landmark_response.landmark_annotations]

    # Perform logo detection
    logo_response = client.logo_detection(image=image)
    logos = [logo.description for logo in logo_response.logo_annotations]

    # Perform image attributes detection
    image_attr_response = client.image_properties(image=image)
    colors = image_attr_response.image_properties_annotation.dominant_colors.colors

    # Perform safe search detection
    safe_search_response = client.safe_search_detection(image=image)
    safe_search = safe_search_response.safe_search_annotation

    # Perform web detection
    web_response = client.web_detection(image=image)
    web_entities = [entity.description for entity in web_response.web_detection.web_entities]

    # Perform object localization
    object_response = client.object_localization(image=image)
    objects = [obj.name for obj in object_response.localized_object_annotations]

    # Perform document text detection
    doc_text_response = client.document_text_detection(image=image)
    doc_texts = [text.description for text in doc_text_response.text_annotations]

    # Compile results
    results = {
        'Labels': labels,
        'Texts (OCR)': texts,
        'Faces': len(faces),
        'Landmarks': landmarks,
        'Logos': logos,
        'Dominant Colors': colors,
        'Safe Search': safe_search,
        'Web Detection': web_entities,
        'Objects': objects,
        'Document Texts': doc_texts
    }

    return results

# Test with a sample image
# https://upload.wikimedia.org/wikipedia/commons/e/e1/Baby_Face.JPG
# http://www.sumit4allphotography.com/wp-content/uploads/2015/04/paris-013.jpg
sample_image_uri = "https://3.bp.blogspot.com/-n8Psu0B27p0/VS4SVPs6k9I/AAAAAAAAADY/tCUm-ZG8lWs/s1600/JPMorgan.jpg"  # Replace with your image URL
analysis_results = analyze_image(sample_image_uri)
for key, value in analysis_results.items():
    print(f"{key}:\n{value}\n\n")
```

```python
from google.cloud import vision

def analyze_image(uri):
    client = vision.ImageAnnotatorClient()
    image = vision.Image()
    image.source.image_uri = uri

    # Perform label detection
    label_response = client.label_detection(image=image)
    labels = [label.description for label in label_response.label_annotations]

    # Perform text detection (OCR)
    text_response = client.text_detection(image=image)
    texts = [text.description for text in text_response.text_annotations]

    # Perform face detection
    face_response = client.face_detection(image=image)
    faces = [{
        'Joy': face.joy_likelihood,
        'Sorrow': face.sorrow_likelihood,
        'Anger': face.anger_likelihood,
        'Surprise': face.surprise_likelihood,
        'Headwear': face.headwear_likelihood,
        'Bounding Poly': face.bounding_poly.vertices,
        'Facial Landmarks': {landmark.type: (landmark.position.x, landmark.position.y, landmark.position.z) for landmark in face.landmarks}
    } for face in face_response.face_annotations]

    # Perform landmark detection
    landmark_response = client.landmark_detection(image=image)
    landmarks = [{ 'Description': landmark.description, 'Location': (landmark.locations[0].latitude, landmark.locations[0].longitude) } for landmark in landmark_response.landmark_annotations]

    # Perform logo detection
    logo_response = client.logo_detection(image=image)
    logos = [logo.description for logo in logo_response.logo_annotations]

    # Perform image attributes detection
    image_attr_response = client.image_properties(image=image)
    colors = [{
        'Color (RGB)': (color.color.red, color.color.green, color.color.blue),
        'Score': color.score,
        'Pixel Fraction': color.pixel_fraction
    } for color in image_attr_response.image_properties_annotation.dominant_colors.colors]

    # Perform safe search detection
    safe_search_response = client.safe_search_detection(image=image)
    safe_search = {
        'Adult': safe_search_response.safe_search_annotation.adult,
        'Spoof': safe_search_response.safe_search_annotation.spoof,
        'Medical': safe_search_response.safe_search_annotation.medical,
        'Violence': safe_search_response.safe_search_annotation.violence,
        'Racy': safe_search_response.safe_search_annotation.racy
    }

    # Perform web detection
    web_response = client.web_detection(image=image)
    web_entities = [entity.description for entity in web_response.web_detection.web_entities]

    # Perform object localization
    object_response = client.object_localization(image=image)
    objects = [{
        'Name': obj.name,
        'Confidence': obj.score,
        'Bounding Poly': obj.bounding_poly.vertices
    } for obj in object_response.localized_object_annotations]

    # Perform document text detection
    doc_text_response = client.document_text_detection(image=image)
    doc_texts_structure = []
    for page in doc_text_response.full_text_annotation.pages:
        for block in page.blocks:
            for paragraph in block.paragraphs:
                for word in paragraph.words:
                    word_text = ''.join([symbol.text for symbol in word.symbols])
                    doc_texts_structure.append(word_text)

    # Compile results
    results = {
        'Labels': labels,
        'Texts (OCR)': texts,
        'Faces': faces,
        'Landmarks': landmarks,
        'Logos': logos,
        'Dominant Colors': colors,
        'Safe Search': safe_search,
        'Web Detection': web_entities,
        'Objects': objects,
        'Document Texts Structure': doc_texts_structure
    }

    return results

# Test with a sample image
sample_image_uri = "https://upload.wikimedia.org/wikipedia/commons/e/e1/Baby_Face.JPG"
analysis_results = analyze_image(sample_image_uri)
for key, value in analysis_results.items():
    print(f"{key}:\n{value}\n\n")

```
